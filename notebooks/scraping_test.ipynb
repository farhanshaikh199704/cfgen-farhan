```python
from fundus import PublisherCollection, Crawler, Sitemap

# initialize a crawler for us/uk based publishers and restrict to Sitemaps only
crawler = Crawler(PublisherCollection.us, PublisherCollection.uk, restrict_sources_to=[Sitemap])

# crawl 1 million articles and print
for article in crawler.crawl(max_articles=1000000):
  print(article)
```

```python
from datetime import datetime, date

def date_filter(article) -> bool:
    """
    Filter to exclude articles published before January 1, 2025.
    Returns True if the article should be included (pass the filter).
    
    Note: For Fundus crawling, returning True means "include this article"
    """
    start_date = date(2025, 1, 1)  # January 1, 2025
    
    if not hasattr(article, 'plaintext'):
        return False
    
    if hasattr(article, 'publishing_date') and article.publishing_date:
        try:
            pub_date = article.publishing_date
            if isinstance(pub_date, datetime):
                pub_date = pub_date.date()
            
            # Return True to include articles published on or after 2025
            return pub_date >= start_date
        except (ValueError, TypeError, AttributeError):
            logger.warning(f"Could not compare publish date: {article.publishing_date}")
            return False
    
    # If no publish date, don't include the article
    return False


for article in crawler.crawl(max_articles=1000000, only_complete=date_filter):
  if date_filter(article):
    print(article.publishing_date)
    print(article.plaintext)
```
